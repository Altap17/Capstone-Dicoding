# Merged df_Log with df_Content

```python
# Merge df_log with df_content to get 'difficulty'
df_merged = pd.merge(
    df_log,
    df_content[["ucid", "difficulty"]],
    on="ucid",
    how="left"
)
df_merged
```

```python
final_features_map = {
    "uuid": "user_id",
    "ucid": "content_topic_id",
    "upid": "problem_question_id",
    "difficulty": "topic_difficulty",
    "total_sec_taken": "time_spent_sec",
    "problem_number": "problem_order",
    "is_correct": "is_correct"
}
df_processed = df_merged[list(final_features_map.keys())].rename(columns=final_features_map)
df_processed
```

```python
# Handle missing difficulty values
df_processed["topic_difficulty"] = df_processed["topic_difficulty"].fillna("unknown")
```

```python
# Convert 'is_correct' boolean to integer (0 or 1)
df_processed["is_correct"] = df_processed["is_correct"].astype(int)
```

```python
# Aggregate features per user per content/topic
df_aggregated = df_processed.groupby(['user_id', 'content_topic_id']).agg(
    total_time_spent=('time_spent_sec', 'sum'),
    avg_time_per_problem=('time_spent_sec', 'mean'),
    total_problems_attempted=('problem_question_id', 'count'),
    total_correct_answers=('is_correct', 'sum'),
    success_rate=('is_correct', 'mean'), # Mean of 0/1 gives success rate
    max_problem_order=('problem_order', 'max') # Max problem number attempted in this topic
).reset_index()
```

```python
# Ordinal encoding for 'topic_difficulty'
difficulty_mapping = {
    'easy': 1,
    'normal': 2,
    'hard': 3,
    'unknown': 0 # Assign 0 for unknown difficulty
}
```

```python
# Map difficulty from df_content based on content_topic_id
df_aggregated['topic_difficulty_encoded'] = df_aggregated['content_topic_id'].map(df_content.set_index('ucid')['difficulty'])
df_aggregated['topic_difficulty_encoded'] = df_aggregated['topic_difficulty_encoded'].map(difficulty_mapping).fillna(0).astype(int)
```

# EXPLORATORY DATA ANALYSISI

```python
df = pd.read_csv(OUTPUT_CLEAN_DATASET)
```

```python
# Map encoded difficulty back to labels for better plot readability
difficulty_labels = {0: 'Unknown', 1: 'Easy', 2: 'Normal', 3: 'Hard'}
df['topic_difficulty_label'] = df['topic_difficulty_encoded'].map(difficulty_labels)
```

```python
# Q1: Efektivitas Belajar: Apakah menghabiskan waktu lebih lama pada satu topik menjamin tingkat kesuksesan (success rate) yang lebih tinggi?
print("\nQ1: Efektivitas Belajar: Apakah menghabiskan waktu lebih lama pada satu topik menjamin tingkat kesuksesan (success rate) yang lebih tinggi?")
plt.figure(figsize=(12, 8))
sns.jointplot(x="total_time_spent", y="success_rate", data=df, kind="reg", height=7, scatter_kws={'alpha':0.3})
plt.suptitle("Total Time Spent vs. Success Rate (with Regression Line)", y=1.02)
```

```python
# Also, a bar plot for average success rate per difficulty
avg_success_by_difficulty = df.groupby('topic_difficulty_label')['success_rate'].mean().reindex(['Easy', 'Normal', 'Hard', 'Unknown'])
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_success_by_difficulty.index, y=avg_success_by_difficulty.values)
plt.title("Average Success Rate by Topic Difficulty")
plt.xlabel("Topic Difficulty")
plt.ylabel("Average Success Rate")
```

```python
# Q3: Pola Perilaku: Apakah ada perbedaan pola waktu pengerjaan antara siswa yang memiliki tingkat kesuksesan tinggi vs rendah?
print("\nQ3: Pola Perilaku: Apakah ada perbedaan pola waktu pengerjaan antara siswa yang memiliki tingkat kesuksesan tinggi vs rendah?")
# Define high vs low success rate groups
df['success_group'] = pd.cut(df['success_rate'], bins=[0, 0.7, 1.0], labels=['Low Success (<70%)', 'High Success (>=70%)'], include_lowest=True)

plt.figure(figsize=(12, 7))
sns.boxplot(x="success_group", y="avg_time_per_problem", data=df)
plt.title("Average Time Per Problem by Success Group")
plt.xlabel("Success Group")
plt.ylabel("Average Time Per Problem (seconds)")
```

```python
# Q4: Skalabilitas Konten: Bagaimana distribusi beban belajar (jumlah soal) di berbagai tingkat kesulitan?
print("\nQ4: Skalabilitas Konten: Bagaimana distribusi beban belajar (jumlah soal) di berbagai tingkat kesulitan?")
plt.figure(figsize=(10, 7))
sns.boxplot(x="topic_difficulty_label", y="total_problems_attempted", data=df, order=['Easy', 'Normal', 'Hard', 'Unknown'])
plt.title("Total Problems Attempted by Topic Difficulty")
plt.xlabel("Topic Difficulty")
plt.ylabel("Total Problems Attempted")
```

```python
print("\n--- 3. Distribution Analysis of Key Features ---")
# Numerical features for distribution analysis
numerical_features = [
    "total_time_spent",
    "avg_time_per_problem",
    "total_problems_attempted",
    "total_correct_answers",
    "success_rate"
]
```

```python
for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.histplot(df[feature], kde=True, bins=30)
    plt.title(f"Distribution of {feature.replace("_", " ").title()}")
    plt.xlabel(feature.replace("_", " ").title())
    plt.ylabel("Frequency")
```

```python
# Distribution of topic_difficulty_encoded
plt.figure(figsize=(8, 5))
sns.countplot(x="topic_difficulty_encoded", data=df)
plt.title("Distribution of Topic Difficulty (Encoded)")
plt.xlabel("Topic Difficulty (Encoded)")
plt.ylabel("Count")
```

```python
# Select only numerical columns for correlation matrix
correlation_matrix = df[numerical_features + ["topic_difficulty_encoded"]].corr()

plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix of Numerical Features")
```

```python
# Total Time Spent vs. Success Rate (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(x="total_time_spent", y="success_rate", data=df, alpha=0.6)
plt.title("Total Time Spent vs. Success Rate")
plt.xlabel("Total Time Spent (seconds)")
plt.ylabel("Success Rate")
```

```python
# Average Time Per Problem vs. Success Rate (Scatter Plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(x="avg_time_per_problem", y="success_rate", data=df, alpha=0.6)
plt.title("Average Time Per Problem vs. Success Rate")
plt.xlabel("Average Time Per Problem (seconds)")
plt.ylabel("Success Rate")
```
